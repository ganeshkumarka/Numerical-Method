```python
def bisect(function : str, a : int, b : int, Epsilon : float) -> float :
    if function(a) * function(b) >= 0:
        return None
    c = None
    while ((b-a) >= Epsilon):
        c = (a + b) / 2
        if function(c) == 0.0:
            break
        if function(c) * function(a) < 0:
            b = c
        else:
            a = c
    return c


def main():
    function = input("Enter the function in terms of x: ")
    a = int(input("Enter the value of a: "))
    b = int(input("Enter the value of b: "))
    Epsilon = float(input("Enter the value of ε: "))
    func = lambda x: eval(function)
    root = bisect(func, a, b, Epsilon)
    if root is not None:
        print("The value of root is : ", "%.4f"%root)
    else:
        print("Can't find root in given interval")

if __name__ == "__main__":
    main()
```

```python
#regular falsi or false position method
def regular_falsi(function : str, a : int, b : int, Epsilon : float) -> float :
    if function(a) * function(b) >= 0:
        return None
    c = None
    while ((b-a) >= Epsilon):
        c = (a*function(b)-b*function(a))/function(b)-function(a)
        if function(c) == 0.0:
            break
        if function(c) * function(a) < 0:
            b = c
        else:
            a = c
    return c


def main():
    function = input("Enter the function in terms of x: ")
    a = int(input("Enter the value of a: "))
    b = int(input("Enter the value of b: "))
    Epsilon = float(input("Enter the value of ε: "))
    func = lambda x: eval(function)
    root = bisect(func, a, b, Epsilon)
    if root is not None:
        print("The value of root is : ", "%.4f"%root)
    else:
        print("Can't find root in given interval")

if __name__ == "__main__":
    main()
```

```python
def gauss_elimination(A, b):
    n = len(A)

    for i in range(n):
        max_row = i
        for k in range(i+1, n):
            if abs(A[k][i]) > abs(A[max_row][i]):
                max_row = k
        A[i], A[max_row] = A[max_row], A[i]
        b[i], b[max_row] = b[max_row], b[i]

        for k in range(i+1, n):
            factor = A[k][i] / A[i][i]
            for j in range(i, n):
                A[k][j] -= factor * A[i][j]
            b[k] -= factor * b[i]

    x = [0] * n
    for i in range(n-1, -1, -1):
        x[i] = b[i] / A[i][i]
        for k in range(i-1, -1, -1):
            b[k] -= A[k][i] * x[i]

    return x

def input_matrix(rows, cols):
    matrix = []
    print("Enter the elements of the matrix:")
    for i in range(rows):
        row = []
        for j in range(cols):
            element = float(input(f"Enter element at position ({i+1},{j+1}): "))
            row.append(element)
        matrix.append(row)
    return matrix


rows = int(input("Enter the number of rows for matrix A: "))
cols = int(input("Enter the number of columns for matrix A: "))

print("For matrix A:")
A = input_matrix(rows, cols)


b = []
print("Enter the elements of the vector b:")
for i in range(rows):
    element = float(input(f"Enter element {i+1}: "))
    b.append(element)

# Apply Gaussian elimination
solution = gauss_elimination(A, b)

# Output the solution
print("\nThe solution is:")
for i in range(len(solution)):
    print(f"x{i+1} = {solution[i]}")
```

```python
#guass siedal method
import numpy as np

def gauss_seidel(A, b, x0=None, tol=0.5, max_iter=100):

    n = len(b)

    # Initialize solution vector
    if x0 is None:
        x0 = np.zeros(n)

    # Iterate until convergence or max iterations reached
    x = x0.copy()
    for iter_count in range(max_iter):
        x_old = x.copy()
        for i in range(n):
            x[i] = (b[i] - np.dot(A[i, :i], x[:i]) - np.dot(A[i, i+1:], x_old[i+1:])) / A[i, i]

        # Check for convergence
        if np.linalg.norm(x - x_old) < tol:
            break

    return x, iter_count + 1

# Example usage:
A = np.array([[3,0.1,-0.2],
              [0.1,7,-0.3],
              [0.3,-0.2,10]
              ], dtype=float)
b = np.array([7.85,-19.3,71.4], dtype=float)

x, iterations = gauss_seidel(A, b)
print("Solution:", x)
print("Number of iterations:", iterations)
```

```python
#gauss jordan
def gauss_jordan(A, b):

    n = len(A)
    augmented_matrix = [row + [bi] for row, bi in zip(A, b)]

    for i in range(n):
        pivot_row = max(range(i, n), key=lambda k: abs(augmented_matrix[k][i]))
        augmented_matrix[i], augmented_matrix[pivot_row] = augmented_matrix[pivot_row], augmented_matrix[i]

        pivot_value = augmented_matrix[i][i]
        for j in range(i, n + 1):
            augmented_matrix[i][j] /= pivot_value

        for k in range(n):
            if k != i:
                factor = augmented_matrix[k][i]
                for j in range(i, n + 1):
                    augmented_matrix[k][j] -= factor * augmented_matrix[i][j]

    x = [row[-1] for row in augmented_matrix]

    return x


def main():
    A = [[3, -0.1, -0.2], [0.1, 7, -0.3], [0.3, -0.2, 10]]
    b = [7.85, -19.3, 71.4]

    solution = gauss_jordan(A, b)
    for i, xi in enumerate(solution):
        print(f"x{i+1} = {round(xi, 4)}")


if __name__ == "__main__":
    main()
```

```python
#newton-forward-interpolation
import math
def ucal(u, p):
    temp = u
    for i in range(1, p):
        temp = temp * (u - i)
    return temp
def fact(n):
    f = 1
    for i in range(2, n + 1):
        f *= i
    return f
def newton_forward_interpolation(x, y, value):
    n = len(x)
    u = (value - x[0]) / (x[1] - x[0])

    # Calculating the forward difference table
    for i in range(1, n):
        for j in range(n - i):
            y[j][i] = y[j + 1][i - 1] - y[j][i - 1]

    # Calculating the interpolated value
    sum = y[0][0]
    for i in range(1, n):
        sum += (ucal(u, i) * y[0][i]) / fact(i)

    return sum
x = [45, 50, 55, 60]
y = [[0 for _ in range(4)] for _ in range(4)]
y[0][0] = 0.7071
y[1][0] = 0.7660
y[2][0] = 0.8192
y[3][0] = 0.8660

value = 52
result = newton_forward_interpolation(x, y, value)
print(f"Value at {value} is {result}")
```

```python
#lagrange-interpolation
import math

def lagrange_interpolation(x_values, y_values, x):
    n = len(x_values)
    result = 0

    for i in range(n):
        term = y_values[i]
        for j in range(n):
            if i != j:
                term *= (x - x_values[j]) / (x_values[i] - x_values[j])
        result += term

    return result

n = int(input("Enter the number of data : "))

x_values = []
y_values = []

for i in range(n):
    print(f"Enter x{i} and y{i} :")
    x, y = map(float, input().split())
    x_values.append(x)
    y_values.append(y)

x = float(input("Enter the value of x to interpolate: "))

result = lagrange_interpolation(x_values, y_values, x)
print(f"The interpolated value at x = {x} is: {result}")
```

```python
#simpson 1/3
import math

def func(x):
    return math.log(x)

def calc_simpson_integral(lower,upper,n):
    h = (upper-lower)/n
    x = []
    y = []

    i = 0
    while i<=n:
        x.append(lower+h*i)
        y.append(func(lower+h*i))
        i+=1


    print(x)
    print(y)

    result = 0
    for i in range(n+1):
        # print(y[i])
        if i==0 or i==n:
            # print(y[i])
            result+=y[i]
        elif i%2==0:
            result+=2*(y[i])
        else:
            result+=4*(y[i])
    result = (h/3)*result

    return result

answer = calc_simpson_integral(4,5.2,6)
print(answer)
```

```python
#simpson 3/8
import math

def func(x):
    return math.log(x)

def calc_simpson_3_8_integral(lower, upper, n):
    if n % 3 != 0:
        print("incorrect interval")

    h = (upper - lower) / n
    x = []
    y = []
    i = 0
    while i <= n:
        x.append(lower + h * i)
        y.append(func(lower + h * i))
        i += 1

    print(x)
    print(y)

    result = 0
    for i in range(n + 1):
        if i == 0 or i == n:
            result += y[i]
        elif i % 3 == 0:
            result += 2 * y[i]
        else:
            result += 3 * y[i]

    result = (3 * h / 8) * result
    return result

answer = calc_simpson_3_8_integral(4, 5.2, 6)
print(answer)
```
